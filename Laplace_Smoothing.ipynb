```python
X_test=pd.read_csv('test.data',delimiter=' ')
X_test.columns=['docId', 'termId', 'Count']
```

```python
def get_prob(dataset,term_id,cls):
  y=dataset[dataset['label']==cls].shape[0]
  dt=dataset[dataset['termId']==term_id]
  x=dt[dt['label']==cls].shape[0]
  return x/y
```

```python
def prior_prob(dataset):
  main_val=list(dataset['label'].unique())
  P=[]
  for i in range(1,21,1):
    prob=dataset[dataset['label']==i].shape[0]/dataset.shape[0]
    P.append(prob)
  return P  
```

```python
def Laplace_smoothing(dataset,term_id,cls):
  y=dataset[dataset['label']==cls].shape[0]
  dt=dataset[dataset['termId']==term_id]
  x=dt[dt['label']==cls].shape[0]
  return (x+1)/(y+21)
```

```python
for i in range(1,21):
  print(Laplace_smoothing(dataset,3,i))
```

```python
X_test['docId'].unique()
```

```python
def naive_bias_classifier(dataset,X):
  Prid=[]
  main_val=list(X.docId.unique())
  P=prior_prob(dataset)
  for val in main_val:
    Post=[]
    if(val<100):
      for cls in range(1,21,1):
        dt=list(X[X['docId']==val]['termId'].unique())
        Likelihood=1
        for term_id in dt:
          Likelihood*=Laplace_smoothing(dataset,term_id,cls)
        prior=P[cls-1]
        Post.append(Likelihood*prior)
      index=Post.index(max(Post))
      Prid.append(index+1)
      print(index+1,'docId is',val)  
  return Prid    
```

```python
Prid=naive_bias_classifier(dataset,X_test)
```

```python
Prid=naive_bias_classifier(dataset,X_test)
```

```python
Y_test=[1 for i in range(99)]
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Prid,Y_test)
```
